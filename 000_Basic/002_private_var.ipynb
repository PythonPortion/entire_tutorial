```python
class MyClass:
    def __init__(self, name):
        self.__private_var = "Private variable"
        self._protected_var = "Protected variable"
        self.public_var = "Public variable"
    
    def get_private_var(self):
        return self.__private_var

obj = MyClass("John")
print(obj.get_private_var()) # Output: Private variable
```

```python
obj.__private_var
```

```python
obj._protected_var
```

```python
obj.public_var
```

1. Private variables are not truely private in Python.
2. They can still be accessed from outside the class using a technique called name mangling.

```python
obj = MyClass("john")
obj._MyClass__private_var
```
