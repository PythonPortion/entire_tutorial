```python
def get_day(n):
    match n:
        case 0:
            return "Monday"
        case 1:
            return "Tuesday"
        case 2:
            return "Wednesday"
        case 3:
            return "Thursday"
        case 4:
            return "Friday"
        case 5:
            return "Saturday"
        case 6:
            return "Sunday"
        case _:
            return "Invalid day"

print(get_day(0))  # Monday
print(get_day(7))  # Tuesday
```

```python
def test(params):
    match params:
        case {'name': name, 'age': age} if age >= 18:
            print(f"Hello {name}, you are {age} years old. adult")
        case {'name': name, 'age': age} if age < 18:
            print(f"Hello {name}, you are {age} years old. child")
        case _:
            print("Invalid input")

test({'name': 'John', 'age': 20})
test({'name': 'Alice', 'age': 15})
test(('name', 'Bob', 'age', 25))
```
