# Single Inheritance

```python
class Parent:
    def parent_method(self):
        print("This is a parent method.")

class Child(Parent):
    def child_method(self):
        print("This is a child method.")

child_instance = Child()
child_instance.parent_method()  # Output: This is a parent method.
child_instance.child_method()   # Output: This is a child method.
```

# Muiltple Inheritance

```python
class Father:
    def father_method(self):
        print("This is a father method.")

class Mother:
    def mother_method(self):
        print("This is a mother method.")


class Kid(Father, Mother):
    def kid_method(self):
        print("This is a kid method.")


kid = Kid()
kid.father_method()
kid.mother_method()
kid.kid_method()
```

```python
class Division:
    def __init__(self, dividend, divisor):
        """
        dividend: The number to be divided.
        divisor: The number by which the dividend is divided.
        """
        self.dividend = dividend
        self.divisor = divisor

    def divide(self):
        if self.divisor != 0:
            return self.dividend / self.divisor
        else:
            return "Error: Division by zero is not allowed."

class Modulus:
    def __init__(self, dividend, divisor):
        """
        dividend: The number to be divided.
        divisor: The number by which the dividend is divided.
        """
        self.dividend = dividend
        self.divisor = divisor

    def modulus(self):
        if self.divisor != 0:
            return self.dividend % self.divisor
        else:
            return "Error: Modulus by zero is not allowed."


class DivMod(Division, Modulus):
    def __init__(self, dividend, divisor):
        self.dividend = dividend
        self.divisor = divisor
    
    def divmod(self):
        division_result = self.divide()
        modulus_result = self.modulus()
        return int(division_result), modulus_result
```

```python
# Division(10, 2).divide()  # Output: 5.0
# Modulus(10, 3).modulus()  # Output: 1
divmod_instance = DivMod(10, 3)
div = divmod_instance.divide()
print(div)  # Output: 3.3333333333333335

mod = divmod_instance.modulus()
print(mod)  # Output: 1

div, mod = divmod_instance.divmod()
print(div, mod)  # Output: 3 1
```

# MRO

```python
class A:
    def show(self):
        print("Class A")

class B(A):
    def show(self):
        print("Class B")
        super().show()

class C(A):
    def show(self):
        print("Class C")
        super().show()

class D(B, C):
    def show(self):
        print("Class D")
        super().show()

d = D()
d.show()
```

```python
class A:
    def show(self):
        print("Class A")

class B(A):
    def show(self):
        print("Class B")


class C(A):
    def show(self):
        print("Class C")


class D(B, C):
    def show(self):
        print("Class D")


d = D()
d.show()
```
