# Multiple Level Inheritance

- A class is derived from another class which is also derived from another class.

```python
class Network:
    def connectivity(self):
        return self.__class__.__name__ + " is connected"

class Network5G(Network):
    def fast_connectivity(self):
        return self.__class__.__name__ + " has fast connectivity"

class Network5G_LTE(Network5G):
    def ultra_fast_connectivity(self):
        return self.__class__.__name__ + " has ultra fast connectivity"

network = Network5G_LTE()
print(network.connectivity())

print(network.fast_connectivity())
print(network.ultra_fast_connectivity())

print("-------------------")
print(Network5G.connectivity(network)) # same as network.connectivity(), but it's obviously that self == network in this case
```

Why 'self.__class__.__name__' is Network5G_LTE in parent class?

- `self` refers to the instance of the class which is 'Network5G_LTE'
- `self.__class__` refers to the class object of the instance which is 'Network5G_LTE'
- `self.__class__.__name__` refers to the name of the class object which is 'Network5G_LTE'
