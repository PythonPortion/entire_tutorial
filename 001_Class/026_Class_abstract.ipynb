- Abstract class is a class that cannot be instantiated, but can be subclassed. 
- It is used to define a common interface for its subclasses, and to provide a default implementation for methods that subclasses can override.
- 抽象类中定义的抽象方法，子类必须实现

```python
from abc import ABC, abstractmethod

class Demo(ABC):
    @abstractmethod
    def method1(self):
        print("This is method 1")

    @abstractmethod
    def method2(self):
        print("This is method 2")


# demo = Demo()
# demo.method1()

class Child(Demo):
    method1 = lambda self: print("This is child method 1")
    method2 = lambda self: print("This is child method 2")

child = Child()
child.method1()
child.method2()
```
