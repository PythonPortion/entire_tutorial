装饰器 把原始类包了一层（代理模式），内部仍然持有原始对象

```python
def decorator_function(T):
    class Wrapper:
        def __init__(self, x):
            self.wrapper = T(x)
        
        def print_name(self):
            print(f"Name: {self.wrapper.name}")
    
    return Wrapper

@decorator_function 
class MyClass:
    def __init__(self, name):
        self.name = name

obj = MyClass("John")
obj.print_name() # Output: Name: John
```

> \`@decorator_function `
> \`class MyClass:`
> ---
> 上面等价于 `MyClass = decorator_function(MyClass)`
> 所以现在的 MyClass 已经不是原来的类了，而是 Wrapper 类。
> ---
> obj = MyClass("John")  实际上是： obj = Wrapper("John")
