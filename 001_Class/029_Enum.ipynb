```python
from enum import Enum
class Color(Enum):
    RED = 1
    GREEN = 2
    BLUE = 3
    GRAT = 3

print(Color.RED)
print(Color.GREEN)
print(Color.BLUE)
print(Color.GRAT)

print(type(Color.RED))

```

从上面输出可以看出 ENUM 允许值重复，但是结果是第一次出现的那个值，本例中是指`Color.BLUE`

添加 unique 修饰符则可以 规避此问题，如下：

```python
from enum import Enum, unique
@unique
class Color(Enum):
    RED = 1
    GREEN = 2
    BLUE = 3
    GRAT = 4

print(Color.RED)
print(Color.GREEN)
print(Color.BLUE)
print(Color.GRAT)

print(Color(1))
```

也可以直接用 ENUM构造

```python
from enum import Enum
Subject = Enum('Subject', 
('Math',
 'Chinese', 
 'English')
 )
print(Subject.Math)
print(Subject.Chinese)
print(Subject.English)

print(Subject(1))
print(Subject['Math'])
print(Subject.Chinese.value)
print(Subject.English.name)
```

还可以使用 迭代循环在 枚举上

```python
for item in Subject:
    print(item.name, item.value)
```
