### 加锁

> 加锁: 确保同一时间只有一个线程访问特定资源或代码段(示例中: print_time); 目的是避免数据不一致或竞争条件。
> 只有获取锁的线程才能执行 print_time, 其他视图获取锁的线程会被阻塞，直到锁被释放
