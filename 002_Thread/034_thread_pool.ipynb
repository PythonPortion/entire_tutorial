```python
import time
import threading
def squre(number):
    sqr = number * number
    time.sleep(1)
    print("Square Current Thread Name:", threading.current_thread().name)
    print("Number:{} Squared:{}".format(number, sqr))

def cube(number):
    cub = number * number * number
    time.sleep(1)
    print("Cube Current Thread Name:", threading.current_thread().name)
    print("Number:{} Cubed:{}".format(number, cub))




```

```python
squre(3)
cube(3)
```

```python
from multiprocessing.dummy import Pool as ThreadPool
from time import sleep
numbers = [1, 2, 3, 4, 5]
pool = ThreadPool(3)
pool.map(squre, numbers)
pool.map(cube, numbers)
pool.close()
```

```python
from concurrent.futures import ThreadPoolExecutor
from time import sleep
numbers = [2,4,6,8,10]
executor = ThreadPoolExecutor(max_workers=4)
t1 = executor.submit(sum, numbers)
t2 = executor.submit(cube, numbers)
print("Thread 1 executed ? :",t1.done())
print("Thread 2 executed ? :",t2.done())
sleep(2)
print("Thread 1 executed ? :",t1.done())
print("Thread 2 executed ? :",t2.done())
```
