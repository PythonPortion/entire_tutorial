## Scheduling Threads using the Timer class

- The `Time` class of the Python threading module allows you to schedule a function to be called after a certain amount of time.

- This class is a subclass of Thread and serves as an example of creating custom threads.

```python
import threading
import time

def schedule_event(name, start):
    now = time.time()
    elapsed = int(now - start)
    print("Elapsed:", elapsed, "Name:", name)

start = time.time()

print("Start:", time.ctime(start))

# Create threads via Timer class

t1 = threading.Timer(2, schedule_event, args=("Thread 1", start))
t2 = threading.Timer(4, schedule_event, args=("Thread 2", start))
t1.start()
t2.start()

t1.join()
t2.join()

end = time.time()

print("End:", time.ctime(end))

print("Total time:", int(end - start), "seconds")

```

## Scheduling Threads using the sched module

```python
import sched
import time

# create a scheduler
scheduler = sched.scheduler(time.time, time.sleep)

start = time.time()
print("Start time:", time.ctime(start))


scheduler.enter(2, 1, schedule_event, argument=("Event 1", start))
scheduler.enter(4, 1, schedule_event, argument=("Event 2", start))

scheduler.run()

end = time.time()
print("End time:", time.ctime(end))
print("Total time:", int(end - start))


```
