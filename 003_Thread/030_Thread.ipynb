```python
import _thread
import time

def print_name(name, *args):
    print(name, *args)

name1 = "1Lx tutorial ... "
name2 = "2Lx tutorial ... "
name3 = "3Lx tutorial ... "

# create and start a new thread to execute the print_name function
_thread.start_new_thread(print_name, (name1, "1", "2"))
_thread.start_new_thread(print_name, (name2, "1", "2", "3"))
_thread.start_new_thread(print_name, (name3, "1", "2", "3", "4"))

time.sleep(1)  # wait for the threads to finish before exiting the main thread
print("Main thread finished.")
```

为什么上述的输出是乱序的？

结论:

    - 线程调度由 OS 决定
    - CPU 时间片随机
    - Python 没有保证顺序
    - _thread 是非常底层的线程库（没有同步机制）

---

为什么输出“乱序”？

    - 多个线程同时抢 stdout（控制台）输出 导致的。
    
    - print() 不是原子操作！
        - print(name, *args)，底层可能是：
            - 先拼接字符串
            - 再写入 stdout buffer
            - 最后 flush 到控制台
            
---

解决方案:

    - 使用锁（Lock）来实现. [也可以使用队列（Queue）]

---


```python
import _thread
import time

lock = _thread.allocate_lock()

def print_name(name, *args):
    with lock:
        print(name, *args)

name1 = "1Lx tutorial ... "
name2 = "2Lx tutorial ... "
name3 = "3Lx tutorial ... "

# create and start a new thread to execute the print_name function
_thread.start_new_thread(print_name, (name1, "1", "2"))
_thread.start_new_thread(print_name, (name2, "1", "2", "3"))
_thread.start_new_thread(print_name, (name3, "1", "2", "3", "4"))

time.sleep(1)  # wait for the threads to finish before exiting the main thread
print("Main thread finished.")
```
