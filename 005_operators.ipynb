```python
a , b = 20, 10
c = 0

print(bin(a))   #0b10100
print(bin(b))   #0b01010
```

```python
c = a & b # bitwise AND operator: returns 1 if both bits are 1, otherwise 0.
print(c) # Output: 0
```

```python
c = a | b # bitwise OR operator: sets each bit to 1 if either bit is 1
print(c) # Output: 30
print(bin(c)) # Output: 0b11110
```

```python
c = a ^ b # bitwise XOR operator: returns a binary number where each bit is 1 if the corresponding bits of the operands are different, and 0 otherwise.
print(c) # Output: 10
print(bin(c)) # Output: 0b1010
```

```python
c = ~a # bitwise NOT operator: flips all the bits in the number
print(c)
print(bin(c))
```

```python
a = 0O20
print(a)
a = 0O30
print(a)
```
