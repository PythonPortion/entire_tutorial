```python
import math
result = math.comb(5,2) 
# 计算从5个物品中选取2个物品的所有可能组合的数量。
# math.comb(n, k)函数返回的是从n个不同元素中选取k个元素的组合数，不考虑元素的顺序。
result
```

```python
result = math.copysign(10, 100) # 用于将第二个参数的符号复制到第一个参数上
result = math.copysign(-10, -1) # 具体来说，math.copysign(x, y) 的作用是返回一个与 x 绝对值相同的数值，但其符号与 y 的符号相同。
result = math.copysign(-10, 1) # 0 的符号与负数的符号相同，所以返回 -0.
print(result)
```

```python
result = math.factorial(5) # 5! = 5 x 4 x 3 x 2 x 1 = 120, 阶乘
print(result)
```

```python
result = 5.0 % 3.0
print(result)
```

```python
result = math.fmod(5,3)
print(result)
```
