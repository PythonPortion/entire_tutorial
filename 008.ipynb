```python
tmp_dict = {'a': 1, 'b': 2, 'c': 3}
for a in tmp_dict:
    print(a)

print("----")

for a in tmp_dict.keys():
    print(a)

print("----")
for a in tmp_dict.values():
    print(a)



print("----")
for a in tmp_dict:
    print(tmp_dict[a])
```
