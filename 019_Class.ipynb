```python
class SmartPhone:
    def __init__(self, brand, price):
        self.brand = brand
        self.price = price
    
    def description(self):
        return f"{self.brand} phone with price {self.price}"

iphone = SmartPhone("Apple", 999)
print(iphone.description())
```

```python
class Desktop:
    def __init__(self):

        self.__sale_price = 10000
    
    def sell(self):
        print("Sold for", self.__sale_price)

    def set_max_price(self, max_price):
        if max_price > 15000:
            self.__sale_price = 15000

dt = Desktop()
dt.__sale_price = 900
dt.sell()
```

```python
class Employee:
    "Common base class for all employees"
    # Class variable: Only accessible to the class and its derived classes
    emp_count = 0

    def __init__(self, name, salary):
        self.name = name
        self.salary = salary
        Employee.emp_count += 1

anki = Employee("Anki", 50000)
print(anki.name)
print(anki.salary)
print(Employee.emp_count)

lenthew = Employee("Lenthew", 60000)
print(lenthew.name)
print(lenthew.salary)
print(Employee.emp_count)
```

```python
# there are two ways to create class methods.
# 1. Using classmethod() Function which transforms an instance method into a class method.
# 2. Using @classmethod Decorator

```

```python
class Employee:
    emp_count = 0

    def __init__(self, name, salary):
        self.name = name
        self.salary = salary
        Employee.emp_count += 1

    def show_count(self):
        print("Total Employee %d" % Employee.emp_count)

    counter = classmethod(show_count)

    @classmethod
    def counter_a(cls):
        print("Total Employee %d" % cls.emp_count)
    
lx = Employee("Lenthew", 60000)
lx2 = Employee("Lenthew", 60000)
lx.show_count()
lx.counter()
Employee.counter()

lx.counter_a()
Employee.counter_a()
```

```python
class Cloth:
    price = 1000

    @classmethod
    def show_price(cls):
        print("The price of the cloth is", cls.price)

Cloth.show_price()
print(Cloth.price)
```
