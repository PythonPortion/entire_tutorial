# Duch-typing in Python

One form of Polymorphism in Python

In Python, duck-typing is a concept of programming where an object's suitability for a certain operation is determined at runtime, rather than at compile-time. This means that the object's type is not checked against a specific type, but rather whether it has the necessary attributes or methods to perform the operation.

主要看他做什么

Suppose there is a bird that walks like a duck, swims like a duck, and quacks like a duck then it probably is a duck.

```python
class Duck:
    def sound(self):
        return "Quack, quack!"

class AnotherBird:
    def sound(self):
        return "I'm similar to a duck"

def make_sound(duck):
    print(duck.sound())

# creating instances
duck = Duck()
anotherBird = AnotherBird()

# calling methods
make_sound(duck)   
make_sound(anotherBird) 
```
