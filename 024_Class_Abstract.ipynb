```python
from abc import ABC, abstractmethod
# print(ABC.__doc__)
# print("---------------------------------")
# print(abstractmethod.__doc__)
class Shape(ABC):
    @abstractmethod
    def draw(self):
        """Draw the shape."""
        return

class Circle(Shape):
    def draw(self):
        super().draw()
        print("Drawing Circle")

class Rectangle(Shape):
    def draw(self):
        super().draw()
        print("Drawing Rectangle")

shapes = [Circle(), Rectangle()]
for shape in shapes:
    shape.draw()

```
