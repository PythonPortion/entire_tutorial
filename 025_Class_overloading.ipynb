Python itself doesn't support method overloading, but we can simulate it using a technique called "multiple inheritance". Here's an example:

```python
class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def sum(self):
        return self.x + self.y
    
    def sum(self, other):
        return self.x + self.y + other.x + other.y


    def __add__(self, other):
        return Vector(self.x + other.x, self.y + other.y)

    def __str__(self):
        return f"Vector({self.x}, {self.y})"

v1 = Vector(1, 2)
v2 = Vector(3, 4)
print(v1 + v2)  # Output: Vector(4, 6)

print(str(v1))  # Output: Vector(1, 2)

print(v1.sum())  # Output: 3
```

If there are same methon names in the class, only the last one will be adopted in the class.

```python
from multipledispatch import dispatch
class Example:
    @dispatch(int, int)
    def add(self, a, b):
        return a + b

    @dispatch(int, int, int)
    def add(self, a, b, c):
        return a + b + c

    @dispatch(str, str)
    def add(self, a, b):
        return a + b

example = Example()
print(example.add(1, 2))        # Output: 3
print(example.add(1, 2, 3))     # Output: 6
print(example.add("Hello, ", "world!"))  # Output: Hello, world!
```
